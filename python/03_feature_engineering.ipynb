```python
df["churn_status"] = df["churn"].map({
    1:"Exited",
    0:"Retained"
})
```

```python
df["customer_status"] = df["active_member"].map({
    1:"Active",
    0:"Inactive"
})
```

```python
def credit(score):
    if score >= 750:
        return "Excellent"
    elif score >= 650:
        return "Good"
    elif score >= 550:
        return "Average"
    else:
        return "Poor"

df["credit_score_category"] = df["credit_score"].apply(credit)
```

```python
df["customer_segment"].value_counts()
```

```python
def segment(balance):
    if balance >= 150000:
        return "VIP"
    elif balance >= 80000:
        return "Premium"
    else:
        return "Regular"

df["customer_segment"] = df["balance"].apply(segment)
```

```python
df["salary_category"] = pd.cut(
    df["estimated_salary"],
    bins=[0,50000,100000,200000],
    labels=[
        "Low Income",
        "Middle Income",
        "High Income"
    ]
)
```
